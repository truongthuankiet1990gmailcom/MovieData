**A Question:** 
What are the primary factors influencing critic and user scores?

**Answering this question will help us:**
- Identify key attributes that contribute to the success or failure of a movie.
- Provide actionable insights for filmmakers to focus on elements that resonate with audiences and critics.
- Guide marketing strategies by highlighting factors that improve reception, such as popular genres or directors.

**How to answer:**
    
- This question seeks to identify the key variables or attributes that impact how critics and users rate movies. Factors may include:
    - **Genre:** Determine if certain genres, such as drama or thriller, are more critically acclaimed.
    - **Production Budget:** Explore whether higher budgets correlate with better ratings.
    - **Release Time:** Assess seasonal trends, such as whether movies released during holidays tend to score higher.
    - **Marketing and Promotion:** Examine if heavily marketed movies receive higher initial ratings.



<h3 id = "preprocessing1">2.1.1 Preprocessing </h3>

```python
import ast
import pandas as pd

data_q1 = data.copy()

# Dealing with the Genre column
data_q1['Genre'] = data_q1['Genre'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) and x.startswith('[') else None)

# Drop rows where Genre could not be converted to a list
data_q1 = data_q1.dropna(subset=['Genre'])

# One-hot encode the Genre column
data_q1 = data_q1.drop('Genre', axis=1).join(data_q1['Genre'].str.join('|').str.get_dummies())

# Split release date into year, month and day
data_q1['Release Date'] = pd.to_datetime(data_q1['Release Date'])
data_q1['Release Year'] = data_q1['Release Date'].dt.year
data_q1['Release Month'] = data_q1['Release Date'].dt.month
# Drop the original Release Date column
data_q1 = data_q1.drop('Release Date', axis=1)

# Calculate the mean score for each movie
data_q1['Mean UserScore'] = data_q1[['Tomatoes UserScore', 'Meta UserScore']].mean(axis=1)
data_q1 = data_q1.drop(['Tomatoes UserScore', 'Meta UserScore'], axis=1)
data_q1['Mean CriticScore'] = data_q1[['Tomatoes CriticScore', 'Metascore']].mean(axis=1)
data_q1 = data_q1.drop(['Tomatoes CriticScore', 'Metascore'], axis=1)

# One-hot encode the Rating column
data_q1 = data_q1.drop('Rating', axis=1).join(data_q1['Rating'].str.get_dummies())

# Drop columns that are not needed
data_q1 = data_q1.drop(['Title', 'Director','Cast', 'Studio'], axis=1)
```

<h3 id = "preprocessing2">2.2.1 Preprocessing </h3>

```python
inflation_df = pd.read_csv('../Data/inflation_rate.csv')
#Get the unit for each year
def get_unit(year):
    try:
        return float(inflation_df[inflation_df['Year'] == year]['Unit'].iloc[0])
    except IndexError:
        return None
```

```python
#Get a copy of data
data_q2 = data.copy()

#Remove unecessary column
data_q2 = data_q2.drop(['Title', 'Cast', 'Director', 'Genre', 'Tomatoes CriticScore', 'Tomatoes UserScore', 'Metascore', 'Meta UserScore',  'Studio'], axis=1)

#Get release year and unit for each year
data_q2['Year'] = pd.to_datetime(data_q2['Release Date']).dt.year
data_q2['Unit'] = data_q2['Year'].apply(get_unit)

#Apply inflation
data_q2['Production Budget'] = data_q2['Production Budget'] * data_q2['Unit']
data_q2['Domestic Gross'] = data_q2['Domestic Gross'] * data_q2['Unit']
data_q2['Worldwide Gross'] = data_q2['Worldwide Gross'] * data_q2['Unit']

#Year and Unit column is not needed anymore
data_q2 = data_q2.drop(['Release Date','Year', 'Unit'], axis=1)

data_q2
```

<h3 id = "preprocessing3">2.3.1 Preprocessing </h3>

```python
sns.set_style("darkgrid")
```

- Separate rows with multiple genres

```python
# Select required columns and create a copy
genre_scores_df = data[['Tomatoes CriticScore', 'Tomatoes UserScore', 'Genre', 
                      'Metascore', 'Meta UserScore', 'Release Date']].copy()

# Change Release Date into year format
genre_scores_df['Release Date'] = pd.to_datetime(genre_scores_df['Release Date']).dt.year

# Explode the Genre column to create separate rows for each genre
genre_scores_df['Genre'] = genre_scores_df['Genre'].str.strip('[]').str.split(',')
genre_scores_df = genre_scores_df.explode('Genre')

# Clean up the genre strings
genre_scores_df['Genre'] = genre_scores_df['Genre'].str.strip().str.strip("'")

# Display the transformed dataframe
genre_scores_df
```

- Put data rows into different "time period" bins

```python
# Create bins for different time periods
bins = [genre_scores_df['Release Date'].min(), 1984, 1994, 2004, 2014, 2024]
labels = ['Pre-1985', '1985-1994', '1995-2004', '2005-2014', '2015-2024']

# Add a new column with the time period labels
genre_scores_df['Time Period'] = pd.cut(genre_scores_df['Release Date'], 
                                      bins=bins,
                                      labels=labels,
                                      include_lowest=True)

genre_scores_df
```

- Group rows by Time Period and Genre and calculate the mean

```python
# Get counts first
genre_counts = genre_scores_df.groupby(['Time Period', 'Genre'], observed=True).size().reset_index(name='Count')

# Calculate averages with renamed columns
avg_genre_score_df = genre_scores_df[['Tomatoes CriticScore', 'Tomatoes UserScore', 'Genre',
                      'Metascore', 'Meta UserScore', 'Time Period']]

avg_genre_score_df = avg_genre_score_df.groupby(['Time Period', 'Genre'], observed=True)[['Tomatoes CriticScore', 'Tomatoes UserScore',
                      'Metascore', 'Meta UserScore']].mean().reset_index()

# Rename the score columns to add (Avg)
avg_genre_score_df = avg_genre_score_df.rename(columns={
    'Tomatoes CriticScore': 'Tomatoes CriticScore (Avg)',
    'Tomatoes UserScore': 'Tomatoes UserScore (Avg)', 
    'Metascore': 'Metascore (Avg)',
    'Meta UserScore': 'Meta UserScore (Avg)'
})

# Merge with counts
avg_genre_score_df = pd.merge(avg_genre_score_df, genre_counts, on=['Time Period', 'Genre'])

avg_genre_score_df
```

- Prepare 5 dataframes for each period

```python
# Split the dataset into separate DataFrames based on Time Period
pre_1985_df = avg_genre_score_df[avg_genre_score_df['Time Period'] == 'Pre-1985']
df_1985_1994 = avg_genre_score_df[avg_genre_score_df['Time Period'] == '1985-1994']
df_1995_2004 = avg_genre_score_df[avg_genre_score_df['Time Period'] == '1995-2004']
df_2005_2014 = avg_genre_score_df[avg_genre_score_df['Time Period'] == '2005-2014']
df_2015_2024 = avg_genre_score_df[avg_genre_score_df['Time Period'] == '2015-2024']
```

<h3 id = "preprocessing5">2.5.1 Preprocessing </h3>

- First we will need to understand the meaning of each rating to have a more comprehensive view of rating system:
    1. **G:** General audiences – All ages admitted.
    2. **PG:** Parental guidance suggested – Some material may not be suitable for children.
    3. **PG-13:** Parents strongly cautioned – Some material may be inappropriate for children under 13.
    4. **R:** Restricted – Under 17 requires accompanying parent or adult guardian.
    5. **NC-17:** Adults only – No one 17 and under admitted.
    6. **TV-G:** Most parents will find this program suitable for all ages.
    7. **TV-PG:** This program contains material that parents may find unsuitable for younger children.
    8. **TV-MA:** This program is specifically designed to be viewed by adults and therefore may be unsuitable for children under 17.
    9. **TV-14:** This program contains material that many parents would find unsuitable for children under 14 years of age.
    10. **Unknown:** This movie or program is not rated due to some reasons

- Secondly, we will need to see the number of movies' ratings over the time 
    - Using the binning method of 10 years to see the overall trend more generally (Because the number of years are too many from 1910 - 2020)

```python
# Number of movies per rating over the years
data_q5 = data.copy()

# Create column for release year
data_q5['Release Year'] = pd.to_datetime(data_q5['Release Date']).dt.year

# Create bin for release year
bin_age = [1910, 1920, 1930, 1940, 1950, 1960, 1970, 1980, 1990, 2000, 2010, 2020, 2024]
label_age = ['1910-1919', '1920-1929', '1930-1939', '1940-1949', '1950-1959', '1960-1969', '1970-1979', '1980-1989', '1990-1999', '2000-2009', '2010-2019', '2020-2024']
data_q5['Release Year Bin'] = pd.cut(data_q5['Release Year'], bins=bin_age, labels=label_age)

data_q5['Release Year Bin']
```

```python
data_q5_1 = data.copy()
# Dealing with the Genre column
data_q5_1['Genre'] = data_q5_1['Genre'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) and x.startswith('[') else None)

# Drop rows where Genre could not be converted to a list
data_q5_1 = data_q5_1.dropna(subset=['Genre'])

# One-hot encode the Genre column
data_q5_1 = data_q5_1.drop('Genre', axis=1).join(data_q5_1['Genre'].str.join('|').str.get_dummies())

# Drop columns that are not needed
data_q5_1 = data_q5_1.drop(['Title', 'Director','Cast', 'Studio', 'Tomatoes CriticScore', 'Tomatoes UserScore', 'Production Budget', 'Domestic Gross', 'Release Date', 'Worldwide Gross', 'Metascore', 'Meta UserScore'], axis=1)

#Get a dataframe consist of
genres = data_q5_1.columns[1:]
ratings = list(data_q5_1['Rating'].unique())
columns = list(data_q5_1['Rating'].unique())
columns.insert(0, 'Gerne')


genre_df = []
for genre in genres:
    genre_row = [genre]
    for rating in ratings:
        genre_data = data_q5_1[data_q5_1[genre] == 1]
        genre_row.append(len(genre_data[genre_data['Rating'] == rating]))
    genre_df.append(genre_row)

genre_df = pd.DataFrame(genre_df, columns=columns)
genre_df.head()
```
